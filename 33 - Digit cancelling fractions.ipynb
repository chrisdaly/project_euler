## Description

The fraction 49/98 is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that 49/98 = 4/8, which is correct, is obtained by cancelling the 9s.

We shall consider fractions like, 30/50 = 3/5, to be trivial examples.

There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.

If the product of these four fractions is given in its lowest common terms, find the value of the denominator.

```python
import itertools
import pandas as pd
```

```python
fractions = []

# Generate all fractions -> 10/11, 10/12 ... 10/19, 11/12, 
for d in range(0, 10):
    for i in range(0, 10):
        for j in range(0, 10):
            tops = list(itertools.permutations([i, d], 2))
            bottoms = list(itertools.permutations([j, d], 2))
            print(tops, bottoms)
            for t in tops:
                for b in bottoms:
                    numerator = int(f"{t[0]}{t[1]}")
                    denominator = int(f"{b[0]}{b[1]}")
                    if (numerator < denominator) and ((numerator % 10 != 0) and (denominator % 10 != 0)):
                        division = numerator / denominator
                        fractions.append({"numerator": numerator, "denominator": denominator, "division": division})
```

```python
df = pd.DataFrame(fractions)
```

```python
df[df["denominator"] == 98]
```
